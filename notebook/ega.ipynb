```python
files = ['analyses.csv',  'analysis_sample.csv',  'dataset.csv',  'experiments.csv',  'run_sample.csv',  'runs.csv', 
         'sample_file.csv',  'samples.csv',  'studies.csv',  'study_analysis_sample.csv',  'study_experiment_run_sample.csv']
```

```python
files
```

```python
for ii_file in files: 
    dataframes = []
    for id in egad:
        fname = id + '/' + ii_file
        df = pd.read_csv(fname)  
        dataframes.append(df)  
        
    combined_df = pd.concat(dataframes, ignore_index=True)

    fname_out = 'combined_' + ii_file
    combined_df.to_csv(fname_out, index=False)
        
```
